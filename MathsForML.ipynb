```python
#a) Plot the  data given
import numpy as np
import matplotlib.pyplot as plt

mean = [0,0]
cov = [[1,0],[0,1]]

x,y = np.random.multivariate_normal(mean, cov, 100).T
plt.plot(x,y, 'X')
plt.axis('equal')
plt.show()
```

```python
#b) The Scatter plot changes when the mean is changed
mean = [-1,1]

x,y = np.random.multivariate_normal(mean, cov, 100).T
plt.plot(x,y, 'X')
plt.axis('equal')
plt.show()
```

```python
#c) The Scatter plot changes when the variance is doubled
mean = [0,0]
cov = [[2,0],[0,2]]

x,y = np.random.multivariate_normal(mean, cov, 100).T
plt.plot(x,y, 'X')
plt.axis('equal')
plt.show()
```

```python
#d) The Scatter plot changes when the Covariance matrix is changed.
mean = [0,0]
cov = [[1,0.5],[0.5,1]]

x,y = np.random.multivariate_normal(mean, cov, 100).T
plt.plot(x,y, 'X')
plt.axis('equal')
plt.show()
```

```python
#e) The Scatter plot changes when the Covariance matrix is changed.
mean = [0,0]
cov = [[1,-0.5],[-0.5,1]]

x,y = np.random.multivariate_normal(mean, cov, 100).T
plt.plot(x,y, 'X')
plt.axis('equal')
plt.show()
```
