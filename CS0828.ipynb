```python
#Solving using sklearn
#importing required libraries
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Creating data with 100 datapoints
x = random.rand(100)
c = random.rand(100)
plt.scatter(x,4*x+c,label='Data')
y= 4*x+c
plt.xlabel("X")
plt.ylabel("Y")

x=np.array(x).reshape((-1,1))
#Fitting the data using linear regression
model = LinearRegression()
model.fit(x,y)
print('The intercept: ',model.intercept_ )
p = model.intercept_
print('The slope: ',model.coef_[0] )
q = model.coef_[0]

plt.plot(x,q*x+p,'c')
plt.show()
```

```python
#Solving using batch gradient descent
#importing required libraries
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
#from sklearn.linear_model import LinearRegression

# Creating data with 100 datapoints
n=100
x = random.rand(n)
c = random.rand(n)
plt.scatter(x,4*x+c,label='Data')
y= 4*x+c
plt.xlabel("X")
plt.ylabel("Y")

#Random values for the slope and the intercept
m = random.random()*10
b = random.random()*10
# value for the learning rate alpha
alpha = 0.3
j=0
# Taking 500 iterations for convergence
for i in range(500):
    j = 0
    for l in range(n):
        #loss function
        j += 1/n*pow(x[1]*m+b-y[1],2)
    delm = 0
    delb = 0
    for k in range(n):
        delm += 1/n*2*(m*x[k]+b-y[k])*x[k]
        delb += 1/n*2*(m*x[k]+b-y[k])
    #updating our slope and intercept
    m = m - alpha*delm
    b = b - alpha*delb
print("The loss function: ",j)
print("THe slope: ",m)
print("The intercept: ",b)
plt.plot(x,x*m+b, 'c')
plt.show()
```

```python
#Solving using mini- batch gradient descent
#importing required libraries
import numpy as np
from numpy import random
import matplotlib.pyplot as plt

n=100
x = random.rand(n)
c = random.rand(n)

plt.scatter(x,4*x+c,label='Data')
y = 4*x+c
plt.xlabel("X")
plt.ylabel("Y")
#Splitting the batch into minibatches each with 10 numbers
x.resize(10,10)
y.resize(10,10)

#Random values for the slope and the intercept
m = random.random()*10
b = random.random()*10
# value for the learning rate alpha
alpha = 0.3
j=0
# Taking 500 iterations for convergence
for i in range(500):
    for z in range(10):
        j = 0
        for l in range(10):
            #the loss function
            j += 1/10*pow(x[z][1]*m+b-y[z][1],2)
        delm = 0
        delb = 0
        for k in range(10):
            delm += 1/10*2*(m*x[z][k]+b-y[z][k])*x[z][k]
            delb += 1/10*2*(m*x[z][k]+b-y[z][k])
    #updating our slope and intercept
    m = m - alpha*delm
    b = b - alpha*delb
    
print("The loss function: ",j)
print("THe slope: ",m)
print("The intercept: ",b)
plt.plot(x,x*m+b, 'c')
plt.show()


```

```python
#Solving using stochastic gradient descent
#importing required libraries
import numpy as np
from numpy import random
import matplotlib.pyplot as plt

n=100
x = random.rand(n)
c = random.rand(n)

plt.scatter(x,4*x+c,label='Data')
y = 4*x+c
plt.xlabel("X")
plt.ylabel("Y")

#Random values for the slope and the intercept
m = random.random()*10
b = random.random()*10
# value for the learning rate alpha
alpha = 0.1
j = 0
# Taking 500 iterations on every point for convergence
for i in range(500):
    for l in range(n):
        j = 0
        #the loss function
        j += pow(x[l]*m+b-y[l],2)
        delm = 0
        delb = 0
        delm += 2*(m*x[l]+b-y[l])*x[l]
        delb += 2*(m*x[l]+b-y[l])
        m = m - alpha*delm
        b = b - alpha*delb
print("The loss function: ",j)
print("THe slope: ",m)
print("The intercept: ",b)
plt.plot(x,x*m+b, 'c')
plt.show()
    
```
